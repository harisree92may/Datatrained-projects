5. Scrape the details of trending repositories on Github.com.

C) Contributors count

D) Language usedASSIGNMENT

Note: - From the home page you have to click on the trending option from Explore menu through code

```python
#lets connect to the webdriver
driver=webdriver.Chrome("chromedriver.exe")
```

```python
url='https://github.com/'
driver.get(url)
```

```python
#click on explore 
explore=driver.find_elements_by_xpath("//ul[@class='d-lg-flex list-style-none']//li[4]//details//summary")
for i in explore:
    if i.text=="Explore":
        i.click()
        break
```

```python
#finding the element for trending button
element=driver.find_elements_by_xpath("//ul[@class='list-style-none mb-3']//li//a")
for i in element:
    i.get_attribute("href")
element    
```

```python
element[3].click()
```

```python
#scrapping the urls
urls=[]
for i in driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']//a"):
    urls.append(i.get_attribute('href'))
urls    
```

```python
#creating empty list
repository_name=[]
repo_des=[]
language_used=[]
contributors_count=[]
for i in urls:
    driver.get(i)
    l=[]
    languages=driver.find_elements_by_xpath("//li[@class='d-inline']//a//span[1]")# fetching languages used
    if languages:
        for i in languages:
            l.append(i.text)
    else:
        l.append("No languages used")
    language_used.append(l)
    time.sleep(4)
# fetching the name of the repository
    try:
        name=driver.find_element_by_xpath("//strong[@class='mr-2 flex-self-stretch']//a")
        repository_name.append(name.text)
    except NoSuchElementException:
        repository_name.append('-')
#fetching description 
    try:
        description=driver.find_element_by_xpath("//div[@class='BorderGrid-cell']//p")
        repo_des.append(description.text)
    except NoSuchElementException:
        repo_des.append('-')

                    
# fetching contributors count
    try:
        count=driver.find_element_by_xpath("//h2[@class='h4 mb-3']/a[contains(text(),'Contributors')]/span")

        contributors_count.append(count.text)

    except NoSuchElementException:
        contributors_count.append('-')
    
```

```python
repository_name
```

```python
repo_des
```

```python
language_used
```

```python
contributors_count
```

```python
#lets check the length of the lists
len(repository_name),len(repo_des),len(language_used),len(contributors_count)
```

```python
#lers create a dataframe
Github_repository=pd.DataFrame(({}))
Github_repository["Repository Name"]=repository_name
Github_repository["Repository Description"]=repo_des
Github_repository["Languages Used"]=language_used
Github_repository["Contributors Count"]=contributors_count
```
